```python
import pandas as pd
from pymongo import MongoClient


csv_file = "Downloads/Inpatient_Rehabilitation_Facility-General_Information_Sep2024 - Inpatient_Rehabilitation_Facility-General_Information_Sep2024.csv - Inpatient_Rehabilitation_Facility-General_Information_Sep2024 - Inpatient_Rehab.csv"
df = pd.read_csv(csv_file)

print(df)
```

```python
client = MongoClient('localhost', 27017)
db = client['hospital_db']
collection = db['rehab_facilities']


records = df.to_dict(orient='records')
collection.insert_many(records)


date_threshold = pd.to_datetime('2011-10-01')
df['Certification Date'] = pd.to_datetime(df['Certification Date'])

non_profit_after_2011 = df[(df['Ownership Type'] == 'Non-profit') & (df['Certification Date'] > date_threshold)]
print("\nNon-Profit providers certified after 10-01-2011:")
print(non_profit_after_2011[['Provider Name', 'Certification Date']])

```

```python
birmingham_profit_providers = df[(df['City/Town'] == 'BIRMINGHAM') & (df['Ownership Type'] == 'Profit')]
print("\nProviders in BIRMINGHAM with Ownership Type PROFIT:")
print(birmingham_profit_providers[['Provider Name', 'City', 'Ownership Type']])
```

```python
zip_range_providers = df[(df['ZIP Code'] >= 85000) & (df['ZIP Code'] <= 90000)]
print("\nProviders with ZIP codes between 85000 and 90000:")
print(zip_range_providers[['Provider Name', 'ZIP Code']])
```

```python
for_profit_count = df[df['Ownership Type'] == 'PROFIT'].shape[0]
non_profit_count = df[df['Ownership Type'] == 'NON-PROFIT'].shape[0]

print(f"\nNumber of For-Profit providers: {for_profit_count}")
print(f"Number of Non-Profit providers: {non_profit_count}")
```
